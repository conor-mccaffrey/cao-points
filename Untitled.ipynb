# PDF: Probability Density Function
***

```python
# Normal Distribution

def normalpdf(x,mu = 0.0, sigma = 1.0):
    """ Calculate the PDF of a normal distribution"""
    numer = np.exp(-0.5 * np.square((x - mu) / sigma))
    denom = sigma * np.sqrt(2.0 * np.pi)
    return (numer / denom)
```

```python
# Test Value.
normalpdf(np.array([-1.0,0.0,1.0]))
```

```python
# Create some x-values
x= np.linspace(-5.0, 5.0, 1000)

# Calculate y-values from x
y = normalpdf(x)


# Create a plot

fig,ax = plt.subplots()

# Plot the values
ax.plot(x,y,'r-')
```

```python
# Create a plot
fig,ax = plt.subplots()

# Create some x-values
x= np.linspace(-5.0, 5.0, 1000)


# Calculate y-values from x with mu = 0.0 and sigma = sqrt(0.2)
y = normalpdf(x, 0.0, np.sqrt(0.2))
# Plot the values
ax.plot(x,y,'b-')

# Calculate y-values from x with mu = 1.0 and sigma = 1.0
y = normalpdf(x)
# Plot the values
ax.plot(x,y,'r-')

# Calculate y-values from x with mu = 0.0 and sigma =sqrt(0.2)
y = normalpdf(x, sigma= np.sqrt(5.0))
# Plot the values
ax.plot(x,y,'y-')

# Calculate y-values from x with mu = 2.0 and sigma =sqrt(0.5)
y = normalpdf(x, -2.0, np.sqrt(0.5))
# Plot the values
ax.plot(x,y,'g-')
```

## Fill Between 

***

```python
# Create some x-values
x= np.linspace(-5.0, 5.0, 1000)

# Calculate y-values from x
y = normalpdf(x)


# Create a plot
fig,ax = plt.subplots()

# Plot the values
ax.plot(x,y,'k-')

# limit x values
x= np.linspace(-2.0, 2.0, 1000)

# Calculate PDF values
y = normalpdf(x)

# Fill between x axis and y - for x values betweeen -2.0 and 2.0
ax.fill_between(x,0,y)

# Draw a couple of vertical lines
plt.axvline(-2.0, color = 'grey', linestyle = ':')
plt.axvline(2.0, color = 'grey', linestyle = ':')

# Draw some text
ax.text(0, 0.2, '~95%', horizontalalignment= 'center')
```

```python
# Probabilty of seeing values between 2.0 and -2.0
stats.norm().cdf(2.0) - stats.norm().cdf(-2.0)
```

```python
# Load iris

df = pd.read_csv('iris.csv')
```

```python
# Extract sepal lengths
vals = df['sepal_length'].to_numpy()[:10]
vals
```

```python
# Extract sepal lengths
vals = df['sepal_length'].to_numpy()[:10]
vals
# Extract sepal lengths
vals = df['sepal_length'].to_numpy()[:10]
vals
# Extract sepal lengths
vals = df['sepal_length'].to_numpy()[:10]
vals

# Minimum and Maximum values
low, high = vals.min(), vals.max()

# Create some space
diff = high - low
low = low - diff
high = high + diff

# Create some x-values
x= np.linspace(low, high, 1000)


# Create a plot
fig,ax = plt.subplots()

# Set the initial previous y values to 0
prev = 0 * x


# Select a value from the data
for i in vals: 
    # Calculate y-values from x
    y = normalpdf(x, mu=i, sigma=0.1 )
    # Add the previous sums of the y values
    y = y + prev
    
    
    # Fill between x axis and yu
    ax.fill_between(x,prev,y, alpha = 0.3)

    # update the prev array
    prev = y
    
# Plot the values
    ax.plot(x,y,'k-')
```
