### Overview of matplotlib.pyplot Python Package

The `matplotlib.pyplot` package is contained within the `matplotlib` library. It is first necessary (and prudent) to understand the mechanics around matplotlib before delving too deeply into matplotlib.pyplot. Matplotlib is a multiplatform data visualisation plotting library used in Python and NumPy [4,5]. It was originally compiled by John Hunter as a patch to IPython but has instead evolved to become the staple-selection for Python plotting [4]. The matplotlib library contains over 70,000 lines of code, however matplotlib's popularity stems from the fact that a few simple lines of code is all that is required to generate a visual data plot [6,5]. Several other notable advantages of matplotlib include [6,7]:
- Ability to control individual elements of a plot i.e. size, legend, axis limits, labels
- High-quality publishable outputs that can be used in scientific journals
- Integrated with <b>LaTeX</b> markup



The following code is used to import `matplotlib.pylot`:
```python
import matplotlib.pyplot as plt
```

Matplotlib contains <b>two</b> dual-APIs (Application Programming Interfaces):
- a MATLAB-style state-based interface (Python API interface)
- an Object-orientated interface

Matplotlib has it's origins as a potential replacement to MATLAB in Python, so naturally inherits a number of characteristics from it. This interface is also known as a <i>stateful</i> interface, in where the current figure and axes are tracked by `pyplot` so that we don't have to [4]. This interface is ideally suited to beginners and simple plots, as not much user input is required in order to create plots. The `matplotlib.pyplot` commands are encompassed in this <i>stateful</i> interface, enabling the opening of figures on-screen.  



````python
# Call default_rng to get a new instance of a Generator, from official documentation [9]
# Creating some data to display Pyplot vs. Object-orientated style
rng = np.random.default_rng() 
arr = rng.integers(0,5, size=150)  # Create some data
arr
````

````python
# Let's create a simple plot using Pyplot-style
mpl.rcParams['lines.linewidth'] = 2 # changing global settings for this case [17]
plt.figure(figsize=(8,6))
plt.plot(arr, label='arr')  # Plot some data on the axes.
plt.plot(arr**2, label='arr$^2$')  # Incorporating LaTeX as already described
plt.plot(arr**3, label='arr$^3$')
plt.xlabel('x')  # Add an x-label to the axes.
plt.ylabel('y') # Add a y-label to the axes.
plt.title('Simple Plot displaying stateful approach')  # Add a title to the axes.
plt.legend(); # Add a legend.
````

The Object-orientated (<i>stateless</i>) interface accounts for occasions when the plot is more complex and requires much more user input and control. The most basic example here would be in the use of <i>subplots</i>. The concept of objects in a matplotlib graphic can be illustrated as such [8]:
- A <i>Figure</i> object acting as the 'container' for the visual
- <i>Axes</i> objects, acting as the plot itself contained within the <i>Figure</i>
- Individual aspects such as legends, text boxes, tick-marks, labels etc. 



<i>Adapted from Reference 8</i>

````python
# Object-orientated style
plt.rcParams["figure.figsize"] = [16,9]
fig, (ax1,ax2) = plt.subplots(1,2)  # Creat figures and axes with pyplot.subplots(), one row and two columns
fig.suptitle('Simple Plots describing OO-style')


# Plot 1
with plt.style.context('seaborn'): #adapted from reference 15
    ax1.plot(arr, label='arr')  # Plot some data on the axes.
    ax1.plot(arr**2, label='arr$^2$')  # Incorporating LaTeX as already described
    ax1.plot(arr**3, label='arr$^3$')
    ax1.set_xlabel('x')  # Add an x-label to the axes.
    ax1.set_ylabel('y')  # Add a y-label to the axes.
    ax1.legend( loc='upper left');


# Plot 2
with plt.style.context('seaborn'):
    ax2.plot(arr, label='arr')  # Plot some data on the axes.
    ax2.plot(arr+10, label='arr+2')  # Incorporating LaTeX as already described
    ax2.plot(arr+20, label='arr+3')
    ax2.set_xlabel('x-points')  # Add an x-label to the axes.
    ax2.set_ylabel('y-points')  # Add a y-label to the axes.
    ax2.legend(loc = 'upper right'); # Add a legend.
````

We can clearly see some differences between these two approaches here, upon inspection of the code. In Object-orientated style, we create a <i>fig</i> and <i>ax</i> instance using `plt.subplots()` (default is 1 row and 1 column [8]). It is clear there is a lot more control using the OO-style approach to plotting, however the code is a little more complex. While the data isn't too aesthetically pleasing, it is there to make a point. In OO-style plotting, you have more control over plotting complex data.


While it may appear that we are quite restricted in scope when using solely the `matplotlib.pyplot` package, this is really not the case. There is a plethora of customisable parameters within the `pyplot` package. For the remainder of this section, I want to highlight a fraction of these parameters.

#### Adjusting parameters of our matplotlib.pyplot plots

When considering points to alter within a plot, the first things many people would think of are the colour of the line, shape of the line or even the style of the line. All these parameters can be modified. Please see the official documentation for a full list of available options [10].

````python
# Lets plot some data and play around
with plt.style.context('seaborn-pastel'):
    plt.figure(figsize=(8,6))
    plt.plot(arr, arr, color='blue', linestyle = 'dashed',  label = 'arr')  # specify color by name
    plt.plot(arr, arr*2, color='#A45A5A', linestyle = ':',lw = 5,label = 'arr*2') # HTML colour codes and setting width
    plt.plot(arr, arr*3, color='grey', marker = 'v', ms= 8, label = 'arr*3' ) # adding marker style and size [12]
    plt.plot(arr, arr*4, c='orange', marker = '2' , label ='arr*4') 
    plt.plot(arr, arr*5, ':r', label ='arr*5') 
    plt.grid() # lets put in gridlines (these can also be altered!)
    plt.xlabel('x-values')
    plt.ylabel('y-values')
    plt.title('Displaying several plots with different appearance')
    plt.legend();
````

From simply looking at the legend in the above figure, you can see how we have drastically altered the appearance of our data using the functionalities avaiable in `pyplot`. Our first set of data is blue and dashed, our second set has a Hex colour [11] and the linestyle is dotted and wide, our third set of data has triangle 'marker caps' and is grey, our fourth set of data is orange and has a 'tri-up' marker and final set has combined these parameters in a single string and is red,dotted. Again, the data isn't aesthetically pleasing but I wanted to incorporate the knowledge gained this semester in using the `numpy.random` module.

Another parameter when dealing with plotting data that we would like the control are the axes limits. This can also be accomplished using `matplotlib.pyplot`.

````python
# Lets plot some data and play around with axes limits
with plt.style.context('dark_background'): # changing background for variety
    plt.figure(figsize=(8,6))
    plt.plot(arr, arr*4, c='blue', marker = '1' , label ='arr*4') 
    plt.plot(arr, arr*5, ':g', label ='arr*5') 
    plt.grid()
    plt.xlabel('x-values')
    plt.ylabel('y-values')
    plt.xlim(-1,5)
    plt.ylim(0,29)
    plt.title('Displaying several plots with different appearance')
    plt.legend();
````

Being able to define the axis limits allows a huge amount of control over the visualisation of the graph in Python. In an interesting quirk, you can display the axis limits in reverse by simply reversing the order of the arguments (i.e in our example, plt.xlim(5,-1)) [4]. Using the method `plt.axis()` allows you to enter values for x- and y- axis in a single line of code [4].

````python
# Further modifying axis variables
# Plot 1
with plt.style.context('seaborn'):
    plt.figure() # using Hex colour for background
    plt.plot(arr, arr*4, c='blue', marker = '1' , label ='arr*4') 
    plt.plot(arr, arr*5, ':g', label ='arr*5') 
    plt.xlabel('x-values')
    plt.ylabel('y-values')
    plt.xlim(5,-1)
    plt.ylim(0,22)
    plt.title('Displaying several plots with different appearance')
    plt.legend();

# Plot 2
with plt.style.context('seaborn'):
    plt.figure() 
    plt.plot(arr, arr*4, c='blue', marker = '1' , label ='arr*4') 
    plt.plot(arr, arr*5, ':g', label ='arr*5') 
    plt.xlabel('x-values')
    plt.ylabel('y-values')
    plt.axis([0,6,0,25])
    plt.title('Displaying several plots with different appearance')
    plt.axvline(x=0, ymin=0, ymax=0.2, ls = ':') # adapted from reference 18
    plt.axvline(x=2, ymin=0, ymax=0.4, ls = ':')
    plt.axvline(x=4, ymin=0, ymax=0.6, ls = ':') # we can add vertical lines to the graphs using pyplot
    plt.legend();
````

When displaying data either in a presentation or perhaps even when publishing data in a journal, it is quite common you want to draw attention to a certain data points or areas. This can be acheived using `matplotlib.pyplot`. 

````python
# Normal distribution function to display text annotation
mu, sigma = 5, 0.5 # mean and standard deviation
s = np.random.default_rng().normal(mu, sigma, 100)

# Plotting the resulting random variables
plt.figure(figsize=(7,6))
count, bins, ignored = plt.hist(s, 20, density=True) # adapted from Reference 13.
# The below lines of code is filling in input values for the probability density as outlined in Reference 13.
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=3, label= 'PDF', color='green') # pdf = Probability Density Function
plt.xlabel('x', fontweight = 'bold')
plt.ylabel('Density', fontweight = 'bold')
plt.text(5.5, .8, r'$\mu=5,\ \sigma=0.5$', fontweight = 'bold')
plt.annotate('mean of 5.0', xy=(5, 1.0), xytext=(5, 1.1),
             arrowprops=dict(facecolor='black', shrink=0.005),
             ) # adapted from reference 14
plt.suptitle('Normal Distribution', fontsize = 15)
plt.legend();
````

#### Box Plots
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.boxplot.html#matplotlib.pyplot.boxplot

Box plots (also referred to as 'Whisker Plots') gives the end-user statistical information about the distribution of numeric data [20]. It is primarily used in the detection of outliers within a given dataset [21]. This is a very useful plot in the case of our Iris datset. According to Official Documention [22]:



'Fliers' in this case clearly denote outliers, they are outside the range of 1.5x the inter-quartile range (IQR). Ina box-plot, the x-axis denotes the data that is to be plotted and visualised while the y-axis shows the frequency distribution [20]. Skewness can also be determined when analysing a boxplot. This is identifed by examing the median line of the boxplot. If this line separates the box in two unequal peices, the data is skewed [23]. If the median line is focussed to the left of the box (or above the median), the data is defined as 'skewed to the right'. If the median line is focussed to the right of the box (or below the median), the data is defined as 'skewed to the left'. Some other points to note for the box-plot are as follows (adapted from reference 21):

- The lower, middle and upper part of the box represents the 25th, 50th, and 75th percentile values respectively.
- The top whisker represents Q3+1.5*IQR
- The bottom whisker represents Q1-1.5*IQR

Boxplots are also a great resoruce for describing the variability of a dateset. As a rule of thumb, the larger the interquartile range, the larger the dataset variability [23]

````python
# Let's construct a boxplot and start analysing
# Let's remove non-numeric data for this boxplot
plt.figure(figsize=(18,10))
with plt.style.context('seaborn'):
    data_updated = data[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
    plt.boxplot(data_updated, labels = ('sepal_length', 'sepal_width', 'petal_length', 'petal_width')) # add labels
    plt.suptitle('Boxplot of Iris Fisher Data Variables') # add title
    plt.xlabel('Variable') # x axis label
    plt.ylabel('cm') # y axis label
````

Boxplots are an extremely effective method of visualising outliers. We can immediately identify datapoints that lie outside the range of 1.5 * interquartile range (IQR), thereby allowing classification as an outlier. We can identify 4 outliers in the sepal_width data. Regarding skewness, I believe there is skewness present in each variable, with the most notable skew present in the petal data. 'Petal_length' could be classed as 'skewed to the right' while 'petal_width' is the same. We have already added some elements, such as axes labels and a boxplot title. Now that we have seen the basic Boxplot, we can begin to custonise various aspects.

````python
# Let's construct a boxplot and start customising
plt.figure(figsize=(18,10))
with plt.style.context('seaborn'):
    # notch chages shape of box at median, vert=0 creates a horizontal plot, patch_artist=True fills in boxplot
    plt.boxplot(data_updated, labels = ('sepal_length', 'sepal_width', 'petal_length', 'petal_width'),
                patch_artist = True ,notch =True, vert = 0) 
    plt.suptitle('Boxplot of Iris Fisher Data Variables') 
    plt.xlabel('cm')
    plt.ylabel('Variable') 
````

From changing only a few parameters, we have transformed this boxplot. It is now horizontal (thanks to `vert=0`). We altered our axis labels to account for this. There is a notch present at the median (`notch=True`), which also confirms confirms our suspicions around skewness. The 'box' is filled in thanks to `patch_artist=True`. Again, we can see out outliers present in the sepal_width data. There is a clear between the width of sepals and petals in this data and also between the length of sepals and petals. There are even more parameters we can alter in boxplots.

````python
# Let's construct a boxplot and start customising
plt.figure(figsize=(18,10))
with plt.style.context('seaborn'):
    # change fliers to green, remove caps at end of whiskers, showmean and meanline set to True
    plt.boxplot(data_updated, labels = ('sepal_length', 'sepal_width', 'petal_length', 'petal_width'),
                patch_artist = True ,notch =True, vert = 0, sym='g', showcaps=False, showmeans=True, meanline=True) 
    plt.suptitle('Boxplot of Iris Fisher Data Variables') 
    plt.xlabel('cm')
    plt.ylabel('Variable') 
````

In this plot, we have edited some more features. We have included a meanline `showmeans=True`and `meanline=True` which displays the meanline across the width of the box. This also helps us display the difference between 'mean' and 'median' in a dataset. The mean of a dataset is the value returned when adding all values and dividing by the number of values. The median is the middle value in a dataset when ordered from highest to lowest [24]. `showcaps=False` results in the horizontal line at the end of the whisker not printing. The fliers have also been decorated. 

````python
# Final adjustment of properties based on reference 25
line_props = dict(alpha=0.7)
boxprops = dict(linestyle='dashdot', linewidth=3, color='blue')
flierprops = dict(marker='o', markersize=10, markerfacecolor = 'y', markeredgecolor='black', )
# Let's plot
plt.figure(figsize=(18,10))
with plt.style.context('seaborn'):
    data_updated = data[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
    # set properties for fliers, box and whiskers. Changed width of boxes.
    plt.boxplot(data_updated, labels = ('sepal_length', 'sepal_width', 'petal_length', 'petal_width'),
               whiskerprops = line_props, boxprops = boxprops, flierprops = flierprops, widths = 0.4 ) 
    plt.suptitle('Boxplot of Iris Fisher Data Variables') # add title
    plt.xlabel('Variable') # x axis label
    plt.ylabel('cm') # y axis label
````

For this boxplot, we have changed quite a number of parameters. We created dict objects for our whiskers, boxes and fliers (adapted from reference 25). Using `whiskerprops`,`boxprops` and `flierprops` we could refer to these dict objects in our code. We also changed the width of the boxes in the plot using `width`. We have also altered the transparency of the whiskers using `alpha`. All these boxplots have shown the wide array of parameters than can be changed/altered and analysed in `matplotlib.pyplot.boxplot`.
